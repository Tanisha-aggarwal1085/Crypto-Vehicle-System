```python
import hashlib
import rsa

# ---------------- VEHICLE DATABASE ----------------
vehicle_db = {}

# ---------------- SHA-256 HASH ----------------
def generate_hash():
    message = input("Enter message: ")
    hash_value = hashlib.sha256(message.encode()).hexdigest()
    print("🔐 SHA-256 Hash:", hash_value)

# ---------------- DIGITAL SIGNATURE ----------------
public_key = None
private_key = None

def generate_keys():
    global public_key, private_key
    public_key, private_key = rsa.newkeys(512)
    print("✅ Key Pair Generated Successfully!")

def sign_message():
    global private_key
    if private_key is None:
        print("❌ Generate keys first!")
        return
    message = input("Enter message to sign: ").encode()
    signature = rsa.sign(message, private_key, 'SHA-256')
    print("✍️ Signature created:", signature)

def verify_signature():
    global public_key
    if public_key is None:
        print("❌ Generate keys first!")
        return
    message = input("Enter original message: ").encode()
    signature = input("Enter signature (in bytes format): ")

    try:
        signature = eval(signature)  # convert string to bytes
        rsa.verify(message, signature, public_key)
        print("✅ Signature is VALID")
    except:
        print("❌ Signature is INVALID")

# ---------------- VEHICLE SYSTEM ----------------
def register_vehicle():
    plate = input("Enter Number Plate: ")
    if plate in vehicle_db:
        print("❌ Vehicle already registered!")
        return
    owner = input("Enter Owner Name: ")
    model = input("Enter Vehicle Model: ")

    vehicle_db[plate] = {
        "owner": owner,
        "model": model
    }
    print("✅ Vehicle Registered Successfully!")

def get_vehicle():
    plate = input("Enter Number Plate: ")
    if plate not in vehicle_db:
        print("❌ Vehicle not found!")
    else:
        print("🚗 Owner:", vehicle_db[plate]["owner"])
        print("🚗 Model:", vehicle_db[plate]["model"])

# ---------------- MENU ----------------
def main():
    while True:
        print("\n====== Cryptography & Vehicle System ======")
        print("1. Generate SHA-256 Hash")
        print("2. Generate Key Pair")
        print("3. Sign Message")
        print("4. Verify Signature")
        print("5. Register Vehicle")
        print("6. Get Vehicle Details")
        print("7. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            generate_hash()
        elif choice == "2":
            generate_keys()
        elif choice == "3":
            sign_message()
        elif choice == "4":
            verify_signature()
        elif choice == "5":
            register_vehicle()
        elif choice == "6":
            get_vehicle()
        elif choice == "7":
            print("Exiting...")
            break
        else:
            print("❌ Invalid choice!")

if __name__ == "__main__":
    main()
```
